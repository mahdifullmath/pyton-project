```python
import numpy as np
import matplotlib.pyplot as plt

# تابع برای حل معادله دیفرانسیل به روش اویلر
def euler_method(f, t0, y0, h, n):
    t = np.zeros(n+1)
    y = np.zeros(n+1)
    t[0] = t0
    y[0] = y0
    for i in range(n):
        t[i+1] = t[i] + h
        y[i+1] = y[i] + h * f(t[i], y[i])
    return t, y

# تابع معادله دیفرانسیل
def ode(t, y):
    return -5 * t * y**2 + 5/t - 1/t**2

# شرایط اولیه
t0 = 1
y0 = 1

# طول گام ها
h1 = 0.19
h2 = 0.21

# تعداد گام ها
n1 = int(10 / h1)
n2 = int(10 / h2)

# حل معادله دیفرانسیل با روش اویلر
t1, y1 = euler_method(ode, t0, y0, h1, n1)
t2, y2 = euler_method(ode, t0, y0, h2, n2)

# جواب دقیق
t_exact = np.linspace(1, 10, 100)
y_exact = 1 / t_exact

# رسم نمودار
plt.figure(figsize=(10, 5))
plt.plot(t1, y1, label=f'Euler Method (h={h1})', marker='o', linestyle='-')
plt.plot(t2, y2, label=f'Euler Method (h={h2})', marker='x', linestyle='-')
plt.plot(t_exact, y_exact, label='Exact Solution', linestyle='--')
plt.xlabel('t')
plt.ylabel('y')
plt.title('Approximate and Exact Solutions')
plt.legend()
plt.grid(True)
plt.show()
```
